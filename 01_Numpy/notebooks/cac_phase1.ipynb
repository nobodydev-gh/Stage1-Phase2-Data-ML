###  Array Reshaping

```python
arr =np.arange(12)
print(arr)

#Reshape 

reshaped = arr.reshape((3,4))
print(reshaped)


#to flattened
flattened = reshaped.flatten()
print(flattened)


#Ravel  (returns view , instead of copy)
raveled = reshaped.ravel()
print(raveled)


#Transpose
transpose = reshaped.T
print('\n Transposed array ', transpose)


```

## Sorting

```python
arr = np.array([4,3,5,6,4,8,9,2,1,7])
print("Sorted Array: ", np.sort(arr))


## 2d array

arr_2d = np.array([[5,3],[2,4],[9,1]])
print("Sorted 2d array by column", np.sort(arr_2d, axis=0))
print("Sorted 2d array by row", np.sort(arr_2d, axis=1))
```

```python
## Filter


numbers = np.arange(1,11)
print(numbers)

even_numbers = numbers[numbers % 2 == 0]
print(even_numbers)



## Filter with Mask

mask = numbers > 5
print("Numbers greater than 5 ", numbers[mask])
```

## Fancy Indexing vs np.where()

```python
indices = [0,2,4]
print(numbers[indices])

where_result = np.where(numbers > 9)
print(where_result)

print(numbers[where_result])
```

```python
condition_array = np.where(numbers > 3 , "true", 'false')
print(condition_array)
```
