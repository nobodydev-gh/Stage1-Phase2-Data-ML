```python
aq=pd.read_csv("air_quality_no2_long.csv")

aq.head()
```

```python
aq=aq.rename(columns={"date.utc" : "datetime"})
aq.head()
```

```python
# converting into datetime object

aq["datetime"] = pd.to_datetime(aq["datetime"])

aq["datetime"]
```

```python
aq["datetime"].min(),aq["datetime"].max()
```

```python
#timedelta
aq["datetime"].min() - aq["datetime"].max()
```

```python
aq["month_name"]=aq["datetime"].dt.month_name()

aq["month_name"].head(3)
```

```python
aq.groupby([aq["datetime"].dt.weekday, "location"])["value"].mean()
```

```python
fig,axs=plt.subplots(figsize=(12,4))

aq.groupby([aq["datetime"].dt.hour,])["value"].mean().plot(kind="bar" ,rot=0 ,ax=axs)

plt.xlabel("Hour of day")
plt.ylabel("No2")
```

```python
aq_p=aq.pivot(index="datetime", columns="location", values="value")
aq_p.head(3)
```

```python
aq_p.index.year,aq_p.index.weekday
```

```python
aq_p["2019-05-20":"2019-05-21"].plot()
```

```python
monthly_max = aq_p.resample("MS").max()
monthly_max
```

```python
aq_p.resample("D").mean().plot(style="-o", figsize=(10,5))
```
